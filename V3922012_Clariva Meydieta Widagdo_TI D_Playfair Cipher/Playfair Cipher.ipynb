```python
def prepare_text(text):
    # Menghapus karakter non-alfabet dan mengganti 'J' dengan 'I'
    text = text.upper().replace('J', 'I')
    text = ''.join(filter(str.isalpha, text))
    return text

def generate_key_matrix(key):
    # Membuat matriks kunci
    key = prepare_text(key)
    alphabet = 'ABCDEFGHIKLMNOPQRSTUVWXYZ'
    key = ''.join(sorted(set(key), key=key.index))
    key = key + ''.join(filter(lambda x: x not in key, alphabet))
    key_matrix = [list(key[i:i+5]) for i in range(0, 25, 5)]
    return key_matrix

def find_positions(matrix, char):
    # Mencari posisi karakter dalam matriks
    for i in range(5):
        for j in range(5):
            if matrix[i][j] == char:
                return i, j

def playfair_encrypt(plain_text, key):
    key_matrix = generate_key_matrix(key)
    plain_text = prepare_text(plain_text)
    cipher_text = []
    for i in range(0, len(plain_text), 2):
        char1, char2 = plain_text[i], plain_text[i+1]
        row1, col1 = find_positions(key_matrix, char1)
        row2, col2 = find_positions(key_matrix, char2)
        if row1 == row2:
            cipher_text.append(key_matrix[row1][(col1 + 1) % 5])
            cipher_text.append(key_matrix[row2][(col2 + 1) % 5])
        elif col1 == col2:
            cipher_text.append(key_matrix[(row1 + 1) % 5][col1])
            cipher_text.append(key_matrix[(row2 + 1) % 5][col2])
        else:
            cipher_text.append(key_matrix[row1][col2])
            cipher_text.append(key_matrix[row2][col1])

    return ''.join(cipher_text)

def playfair_decrypt(cipher_text, key):
    key_matrix = generate_key_matrix(key)
    cipher_text = prepare_text(cipher_text)
    plain_text = []
    for i in range(0, len(cipher_text), 2):
        char1, char2 = cipher_text[i], cipher_text[i+1]
        row1, col1 = find_positions(key_matrix, char1)
        row2, col2 = find_positions(key_matrix, char2)
        if row1 == row2:
            plain_text.append(key_matrix[row1][(col1 - 1) % 5])
            plain_text.append(key_matrix[row2][(col2 - 1) % 5])
        elif col1 == col2:
            plain_text.append(key_matrix[(row1 - 1) % 5][col1])
            plain_text.append(key_matrix[(row2 - 1) % 5][col2])
        else:
            plain_text.append(key_matrix[row1][col2])
            plain_text.append(key_matrix[row2][col1])

    return ''.join(plain_text)

# Main program
while True:
    choice = input("Pilih enkripsi (E) atau dekripsi (D): ").upper()
    if choice not in ['E', 'D']:
        print("Pilihan tidak valid.")
        continue

    key = input("Masukkan kunci: ")
    text = input("Masukkan teks: ")

    if choice == 'E':
        result = playfair_encrypt(text, key)
        print("Hasil enkripsi:", result)
    else:
        result = playfair_decrypt(text, key)
        print("Hasil dekripsi:", result)

    another = input("Lanjutkan (Y/N)? ").upper()
    if another != 'Y':
        break
```
