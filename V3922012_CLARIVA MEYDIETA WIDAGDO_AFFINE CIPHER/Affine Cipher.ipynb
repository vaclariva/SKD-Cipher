```python
# Fungsi untuk menghitung modulus inverse a modulo m.
def mod_inverse(a, m):
    for x in range(1, m):
        if (a * x) % m == 1:
            return x
    return None

# Fungsi untuk melakukan enkripsi menggunakan affine cipher.
def affine_encrypt(plain_text, key_a, key_b):
    encrypted_text = ""
    for char in plain_text:
        if char.isalpha():
            char_idx = ord(char) - ord('A')  # Mengubah huruf menjadi indeks (A=0, B=1, ...).
            encrypted_idx = (key_a * char_idx + key_b) % 26  # Rumus enkripsi affine.
            encrypted_char = chr(encrypted_idx + ord('A'))  # Mengubah indeks kembali menjadi huruf.
            encrypted_text += encrypted_char
        else:
            encrypted_text += char  # Menambahkan karakter selain huruf tanpa perubahan.
    return encrypted_text

# Fungsi untuk melakukan dekripsi menggunakan affine cipher.
def affine_decrypt(encrypted_text, key_a, key_b):
    decrypted_text = ""
    key_a_inv = mod_inverse(key_a, 26)  # Menghitung modulus inverse untuk kunci a.
    if key_a_inv is None:
        return "Kunci tidak valid."  # Memeriksa apakah kunci a memiliki modulus inverse.

    for char in encrypted_text:
        if char.isalpha():
            char_idx = ord(char) - ord('A')  # Mengubah huruf menjadi indeks (A=0, B=1, ...).
            decrypted_idx = (key_a_inv * (char_idx - key_b)) % 26  # Rumus dekripsi affine.
            decrypted_char = chr(decrypted_idx + ord('A'))  # Mengubah indeks kembali menjadi huruf.
            decrypted_text += decrypted_char
        else:
            decrypted_text += char  # Menambahkan karakter selain huruf tanpa perubahan.
    return decrypted_text

# Loop utama untuk mengatur pilihan enkripsi atau dekripsi.
while True: #jika benar
    #deskripsi opsi 1 untuk enkripsi dan 2 untuk deskripsi
    choice = input("Pilih operasi:\n1. Enkripsi\n2. Dekripsi\nPilihan (1/2): ")
    #jika inputan selain angka 1 dan 2 maka akan muncul peringatan tidak valid
    if choice not in ['1', '2']:
        print("Pilihan tidak valid. Silakan pilih 1 atau 2.")
        continue

    try: #user input nilai key berupa bilangan bulat yang akan digunakan dalam perhitungan
        key_a = int(input("Masukkan nilai kunci a (bilangan bulat): "))
        key_b = int(input("Masukkan nilai kunci b (bilangan bulat): "))
    # fungsi akan error bila tidak memasukkan nilai dengan selain bilangan bulat
    except ValueError:
        print("Kunci harus berupa bilangan bulat.")
        continue
    #jika user pilih 1
    if choice == '1':
        #user input teks yang ingin dideskripsi
        plain_text = input("Masukkan teks yang ingin dienkripsi: ").upper()
        encrypted_text = affine_encrypt(plain_text, key_a, key_b)
        print("Teks terenkripsi:", encrypted_text)
    #jika user pilih 2
    elif choice == '2':
        #user input teks yang ingin dienkripsi
        encrypted_text = input("Masukkan teks yang ingin didekripsi: ").upper()
        decrypted_text = affine_decrypt(encrypted_text, key_a, key_b)
        print("Teks terdekripsi:", decrypted_text)
    
    #bentuk perulangan
    another = input("Lakukan operasi lainnya? (y/n): ")
    if another.lower() != 'y':
        break
```
